```python
import seaborn as sns
```

```python
# loading the test and train dataset
train_df = pd.read_excel('Data_Train.xlsx')
test_df = pd.read_excel('Test_set.xlsx')
```

```python
# combined train and test dataset (Kaggle competetion)
big_df = train_df.append(test_df, sort=False)
```

## Feature Engineering

```python
big_df['Date'] = big_df['Date_of_Journey'].str.split('/').str[0]
big_df['Month'] = big_df['Date_of_Journey'].str.split('/').str[1]
big_df['Year'] = big_df['Date_of_Journey'].str.split('/').str[2]
# You can also convert the column into datetime object and then extract date, month, year
# big_df['Date'] = pd.to_datetime(big_df['Date_of_Journey'], format='%d/%m/%y').dt.day
# big_df['Month'] = pd.to_datetime(big_df['Date_of_Journey'], format='%d/%m/%y').dt.month
# big_df['Year'] = pd.to_datetime(big_df['Date_of_Journey'], format='%d/%m/%y').dt.year
```

```python
big_df.head()
```

```python
# convert the date, month, year into int data type
big_df['Date']=big_df['Date'].astype(int)
big_df['Month']=big_df['Month'].astype(int)
big_df['Year']=big_df['Year'].astype(int)
```

```python
# dropped the date of journe column after extracting date, month, year from that column
big_df=big_df.drop(['Date_of_Journey'], axis=1)
big_df.head()
```

```python
big_df['Arrival_Time']=big_df['Arrival_Time'].str.split(' ').str[0]
big_df.head()
# You can also convert the Arrival_Time column into datetime object and then extract Hour, Minute
# big_df['Hour'] = pd.to_datetime(big_df['Arrival_Time']).dt.Hour
# big_df['Minute'] = pd.to_datetime(big_df['Arrival_Time']).dt.Minute
```

```python
# checking the unique values in the Total_Stops column
big_df['Total_Stops'].value_counts()
```

```python
# Selecting the null value in the Total_Stops column
big_df[big_df['Total_Stops'].isnull()]
```

```python
# only 1 value missing; lets replace with 1 stop value inside it
big_df['Total_Stops'] = big_df['Total_Stops'].fillna('1 stop')
```

```python
# checking the Total_Stops column again. null value should not appear
big_df[big_df['Total_Stops'].isnull()]
```

```python
# heatmap to visualize the null columns
sns.heatmap(big_df.isnull(), yticklabels=False, cbar=False)
```

```python
# Replaced non-stop with 0 stop value to later extract the numerical values:
big_df['Total_Stops']=big_df['Total_Stops'].replace('non-stop', '0 stop')
big_df.head()
```

```python
# extract the numerical values from Total_Stops column
big_df['Stop'] = big_df['Total_Stops'].str.split(' ').str[0]
big_df.head()
```

```python
# change into integer data type and dropping the Total_Stops column
big_df['Stop'] = big_df['Stop'].astype(int)
big_df=big_df.drop(['Total_Stops'], axis=1)
```

```python
big_df.head()
```

```python
# Extracting the Hour and Minute information from the Arrival_Time column
big_df['Arrival_Hour'] = big_df['Arrival_Time'].str.split(':').str[0]
big_df['Arrival_Minute'] = big_df['Arrival_Time'].str.split(':').str[1]
```

```python
# Converting the extracted new column into  integer data type
big_df['Arrival_Hour'] = big_df['Arrival_Hour'].astype(int)
big_df['Arrival_Minute'] = big_df['Arrival_Minute'].astype(int)
```

```python
# Dropping the column Arrival_Time after extracting the hour and minute information
big_df= big_df.drop(['Arrival_Time'], axis=1)
```

```python
big_df.head()
```

```python
# like departure time , repeating for arrrival time
big_df['Departure_Hour'] = big_df['Dep_Time'].str.split(':').str[0]
big_df['Departure_Minute'] = big_df['Dep_Time'].str.split(':').str[1]
big_df.head()
```

```python
# Departure Hour and Departure Minute column as integer data type
big_df['Departure_Hour'] = big_df['Departure_Hour'].astype(int)
big_df['Departure_Minute'] = big_df['Departure_Minute'].astype(int)
```

```python
# dropped the Dep_Time column after extracting the Departure Hour and Departure Minute
big_df = big_df.drop(['Dep_Time'],axis=1)
big_df.head()
```

```python
# splitting the route column
big_df['Route_1'] = big_df['Route'].str.split('→ ').str[0]
big_df['Route_2'] = big_df['Route'].str.split('→ ').str[1]
big_df['Route_3'] = big_df['Route'].str.split('→ ').str[2]
big_df['Route_4'] = big_df['Route'].str.split('→ ').str[3]
big_df['Route_5'] = big_df['Route'].str.split('→ ').str[4]
```

```python
big_df.head()
```

```python
# visualize the underline distribution
big_df['Price'].plot.kde()
```

```python
# dealing with null values
big_df['Price'].fillna((big_df['Price'].mean()), inplace=True)
```

```python
big_df.head()
```

```python
# heatmap visualization for null values
sns.heatmap(big_df.isnull(), yticklabels=False, cbar=False)
```

```python
big_df['Route_1'].fillna("None", inplace=True)
big_df['Route_2'].fillna("None", inplace=True)
big_df['Route_3'].fillna("None", inplace=True)
big_df['Route_4'].fillna("None", inplace=True)
big_df['Route_5'].fillna("None", inplace=True)
```

```python
big_df.head()
```

```python
# Heatmap to visualize null values
sns.heatmap(big_df.isnull(), yticklabels=False, cbar=False)
# no missing values now
```

```python
# dropping the route column as destination wise route has been extracted
big_df = big_df.drop(['Route'], axis=1)
big_df.head()
```

```python
big_df = big_df.drop(['Duration'], axis=1)
big_df.head()
```

```python
# dealing with ordinal categorical variable by applying Label Encoding class
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
big_df['Airline'] = encoder.fit_transform(big_df['Airline'])
big_df['Source'] = encoder.fit_transform(big_df['Source'])
big_df['Destination'] = encoder.fit_transform(big_df['Destination'])
big_df['Additional_Info'] = encoder.fit_transform(big_df['Additional_Info'])
big_df['Route_1'] = encoder.fit_transform(big_df['Route_1'])
big_df['Route_2'] = encoder.fit_transform(big_df['Route_2'])
big_df['Route_3'] = encoder.fit_transform(big_df['Route_3'])
big_df['Route_4'] = encoder.fit_transform(big_df['Route_4'])
big_df['Route_5'] = encoder.fit_transform(big_df['Route_5'])
```

```python
big_df.head()
```

## Feature Selection

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
df_train = big_df[0:10683]
df_test = big_df[10683:]
```

```python
x = df_train.drop(['Price'], axis=1)
y = df_train['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
# SelectFromModel will select the best features while applying the model
model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
```

```python
model.fit(x_train, y_train)
```

```python
# to see which features got selected into the model
model.get_support()
```

```python
# selected features into the model
selected_features = x_train.columns[(model.get_support())]
selected_features
```

```python
# to compare the column name from above selected features
x_train.columns
```

```python
x_train = x_train.drop(['Year'], axis=1)
```

```python
x_test = x_test.drop(['Year'], axis=1)
```

## RandomForestRegressor

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# number of decision trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# features for split
features = ['auto', 'sqrt']
# level depth in the trees
depth = [int(x) for x in np.linspace(5,30, num = 6)]
# number of samples splitting the node
samples_split = [2, 5, 10, 15, 100]
# sample at each leaf node
sample_leaf = [1, 2, 5, 10]
```

```python
# Grid creation for tuning the hyperparameters

random_grid = {'n_estimators': n_estimators,
              'max_features': features,
               'max_depth': depth,
               'min_samples_split': samples_split,
               'min_samples_leaf': sample_leaf}


print(random_grid)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, scoring= 'neg_mean_squared_error', n_iter = 50, cv = 5, verbose =2, random_state=42)
```

```python
rf_random.fit(x_train, y_train)
```

```python
y_pred = rf_random.predict(x_test)
```

```python
import seaborn as sns

sns.distplot(y_test-y_pred);
```

```python
plt.scatter(y_test, y_pred)
plt.title('Predicted vs. Real Price Comparison')
plt.xlabel('Predicted Price')
plt.ylabel('Given Price')
plt.show();
# you can see a linear relationship between the prediction price values and given price values
```
